```python
import glob
import os
import mysql.connector
import random
import datetime

corporations = {'INTC': 'Intel Corporation', 'AMD':'Advanced Micro Devices', 'AAPL':'Apple', 'FB':'Meta Platforms, Inc.'}
mysql_config =  {
  'user': 'root',
  'password': 'admin321',
  'host': '127.0.0.1'
}
conn = mysql.connector.connect(**mysql_config)
curr = conn.cursor(buffered=True)
curr.execute(f'USE {db_name}')
sql = '''
SELECT symbol,trade_date,close_price - open_price as diff,close_price 
FROM security_master 
WHERE symbol = 'AMD'; '''
curr.execute(sql)
rows = curr.fetchall()
up_days = 0
down_days = 0
tot_val = 0
tot_shares = 0
share_size = 10000
for row in rows:
    symbol = row[0]
    y,m,d = str(row[1].year), ('00'+str(row[1].month))[-2:], ('00'+str(row[1].day))[-2:]
    transaction_id = f'{y}{m}{d}_{row[0]}_{str(random.randint(100,1000))}'
    trade_date = row[1]    
    '''
    if closed up for 5 consecutive days SELL 10000 shares
    sell if down for 5 consecute days '''
    if row[2] > 0:
        up_days += 1
        down_days = 0
    else:
        up_days = 0
        down_days += 1
    if up_days >= 5:
        down_days = 0
        '''SELL signal'''
        if tot_shares > 0:
            factor = 2 if tot_shares - (share_size*2) > 0 else 1
            tot_val += ((share_size*factor) * row[3])
            tot_shares -= (share_size*factor)
            # = YYYYMMDD_symbol_random number, 100 up to 999<br>
            print(f'****** SELL HERE {row[0]} {row[1]} {row[3]}  - {tot_shares} {tot_val}') 
            sql = f'''INSERT INTO security_log
            (transaction_id,trade_date,symbol,trans_type,shares,cost,balance)
            VALUES 
            ('{transaction_id}','{trade_date}','{symbol}','SELL',
            {tot_shares},{round(share_size*row[3],2)},{round(tot_val,2)})
            '''
            print(sql)
            curr.execute(sql)
            conn.commit()  
    if down_days >= 5:
        up_days = 0
        down_days = 0
        '''buy signal'''
        tot_val += (-1 * share_size * row[3])
        tot_shares += share_size
        print(f'****** BUY HERE {row[0]} {row[1]} {row[3]} - {tot_shares} {tot_val}')        
        sql = f'''INSERT INTO security_log
        (transaction_id,trade_date,symbol,trans_type,shares,cost,balance)
        VALUES 
        ('{transaction_id}','{trade_date}','{symbol}','BUY',
        {tot_shares},{round(-1 *share_size*row[3],2)},{round(tot_val,2)})
        '''
        print(sql)
        curr.execute(sql)
        conn.commit()  

print(tot_val, tot_shares)
conn.close()
```

```python
import glob
import os
import mysql.connector

corporations = {'INTC': 'Intel Corporation', 'AMD':'Advanced Micro Devices', 'AAPL':'Apple', 'FB':'Meta Platforms, Inc.'}
mysql_config =  {
  'user': 'root',
  'password': 'admin321',
  'host': '127.0.0.1'
}
conn = mysql.connector.connect(**mysql_config)
curr = conn.cursor(buffered=True)
curr.execute(f'USE {db_name}')
sql = '''
SELECT max(trade_date),min(trade_date) FROM security_master
WHERE year(trade_date) = '2020' and symbol = 'FB' 
; '''
print(sql)
curr.execute(sql)
rows = curr.fetchall()
for row in rows:
    buy_date = row[1]
    sell_date = row[0]
    sql = f'''
    SELECT close_price FROM security_master
    WHERE trade_date = '{buy_date}' and symbol = 'FB' 
    ; '''
    print(sql)
    curr.execute(sql)
    rows = curr.fetchall()
    bought = 0
    sold = 0
    for row in rows:
        bought = -1 * row[0] * 100000
        print(f'Bought price {bought}')
        
    sql = f'''
    SELECT close_price FROM security_master
    WHERE trade_date = '{sell_date}' and symbol = 'FB' 
    ; '''
    print(sql)
    curr.execute(sql)
    rows = curr.fetchall()
    for row in rows:
        
        sold = row[0] * 100000     
        print(f'Sold price {sold}')
        PnL = sold + bought
        print(f'Profit or loss = {PnL}')
conn.close()
```
